```python
from luna.data_processing.utils import *
from luna.models import xgb_clf
```

```python
df_train = load_data('./example_data/classification/train.csv', 'csv')
df_train
```

```python
target = 'rainfall'
```

```python
df_train = preprocess_data(df_train, target)
```

```python
df_train 
```

```python
X_train, X_val, y_train, y_val = split_data(df_train, target)
```

```python
xgb_clf = xgb_clf()
xgb_clf.fit(X_train, y_train)
```

```python
xgb_clf.evaluate(X_val, y_val)
```

```python
# 配置日志系统
import logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)

# 然后再调用评估函数
xgb_clf.evaluate(X_val, y_val)
```
